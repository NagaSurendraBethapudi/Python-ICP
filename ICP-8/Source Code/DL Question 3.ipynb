```python
#Loading the Libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#Importing the data
df = pd.read_csv('Breas Cancer.csv')
```

```python
#Finding the overal info about the data
df.info()
```

##### Observations:
1. There are 32 columns
2. Only one column is categorical (diagnosis)

```python
#Printing the first five rows
df.head()
```

```python
#Finding the unique value or diagnosis column since it is only categorical value in data
df['diagnosis'].unique()
```

```python
#Finding any values in Unnamed: 32 column
df['Unnamed: 32'].unique()
```

#### As we could see there are no values in Unnamed: 32 columns, so we can skip this feature

```python
df = df.drop(['Unnamed: 32'],axis = 1)
```

```python
#Converting the categorical data to int
convert = {"diagnosis": {"M": 0, "B": 1}}
cancer_data = df.replace(convert)
cancer_data.head()
```

```python
#Copying the remaining features other than diagnosis to x
x = cancer_data.iloc[:,2:31]
```

```python
#Copying the diagnosis feature(Predictable feature)
y = cancer_data['diagnosis']
```

```python
#splitting the data into training and testing categories
X_train, X_test, Y_train, Y_test = train_test_split(x, y,test_size=0.25, random_state=87)
```

```python
#Building the sequential model
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=29, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer

#Performing compilation
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,initial_epoch=0)

#Printing Summary and Accuracy
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```

#### Observations  :
1. Loss     = 25%
2. Accuracy = 93%

## Noramalizing the data using standard scalar

```python
#Copying the remaining features other than diagnosis to x
x = cancer_data.iloc[:,2:31]
```

```python
#Normalising the data using Standard scaler
SC = StandardScaler()
SC.fit(x) #Fitting the data
x_normalization = SC.transform(x)
```

```python
#Copying the diagnosis feature(Predictable feature)
y = cancer_data['diagnosis']
```

```python
#splitting the data into training and testing categories
X_train, X_test, Y_train, Y_test = train_test_split(x_normalization, y,test_size=0.25, random_state=87)
```

```python
#Building the sequential model
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=29, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer

#Performing compilation
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,initial_epoch=0)

#Printing Summary and Accuracy
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```

#### Observations  after normalizing the data:
1. Loss     = 14%
2. Accuracy = 97%

<h4> Accuracy was increased and loss of data was decreased after Normalizing the data </h4>
